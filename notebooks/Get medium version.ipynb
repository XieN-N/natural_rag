```python
from pathlib import Path
import yaml
import random
import shutil
```

```python


questions = yaml.safe_load(open('datasets/bl/questions.yaml'))
corpus_info = yaml.safe_load(open('datasets/bl/corpus_info.yaml'))
questions_small = yaml.safe_load(open('datasets/bl_tiny/questions.yaml'))
```

```python
questions_small_texts = [q['text'] for q in questions_small]
questions_texts = [q['text'] for q in questions]
```

```python
random.seed(0)
random.shuffle(questions_texts)
random.shuffle(corpus_info)
```

```python
selected_questions = set(questions_texts[:30] + questions_small_texts)
len(selected_questions)
```

```python
questions_to_save = [x for x in questions if x['text'] in selected_questions]
```

```python
docs = set[str]()
for q_info in questions_to_save:
    for relevant in q_info.get('relevant', []):
        docs.add(relevant['doc_id'])
len(docs)

docs |= set([x['id'] for x in corpus_info][:50])
len(docs)
```

```python
corpus_info_to_save = [x for x in corpus_info if x['id'] in docs]
```

```python
dir = Path('datasets/bl_medium')
dir.mkdir(exist_ok=True, parents=True)

(dir / 'corpus_info.yaml').write_text(yaml.dump(corpus_info_to_save))
(dir / 'questions.yaml').write_text(yaml.dump(questions_to_save))
```

```python
(dir / 'sources').mkdir(exist_ok=True, parents=True)
(dir / 'docs').mkdir(exist_ok=True, parents=True)

for doc_info in corpus_info_to_save:
    doc_name = doc_info["title"] + '.md'
    source_name = doc_info["title"] + doc_info["source_ext"]
    shutil.copy(f'datasets/bl/sources/{source_name}', dir / f'sources/{source_name}')
    shutil.copy(f'datasets/bl/docs/{doc_name}', dir / f'docs/{doc_name}')
```
