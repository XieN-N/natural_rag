```python
import os
from pathlib import Path
from typing import cast

from openai import OpenAI

from natural_rag.baseline import BUILD_PROMPT, EntriesList

```

```python
from bigtree import Node

a = Node("a")
b = Node("b", parent=a)
c = Node("c", parent=b)
d = Node("d", parent=b)
a.show()
```

```python
client = OpenAI(base_url=os.environ['VSEGPT_BASE_URL'], api_key=os.environ['VSEGPT_KEY'])
```

```python
doc = Path('datasets/bl_small/docs/Decay.md').read_text()
doc = doc.split('Preventing Decay')[0]

completion = client.beta.chat.completions.parse(
    model="openai/gpt-5-nano",  # Ensure model supports Structured Outputs
    messages=[{"role": "user", "content": BUILD_PROMPT.format(document=doc)}],
    # response_format=EntriesList,
    max_tokens=99999,
    max_completion_tokens=99999,
)
```

```python
print(completion.choices[0].message.model_dump_json(indent=4, ensure_ascii=False))
```
